# Heart Disease Prediction using Neural Networks

```python
import sys
import pandas as pd
import numpy as np
import sklearn
import matplotlib
import keras

import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
```

###  Importing the Dataset

```python
# import the heart disease dataset
url = "http://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.cleveland.data"
names = ['age',
        'sex',
        'cp',
        'trestbps',
        'chol',
        'fbs',
        'restecg',
        'thalach',
        'exang',
        'oldpeak',
        'slope',
        'ca',
        'thal',
        'class']
cleveland = pd.read_csv(url, names=names)
```

```python
# remove missing data indicated with a "?"
data = cleveland[~cleveland.isin(['?'])]
data.tail()
```

```python
# drop rows with NaN values
data = data.dropna(axis=0)
data.tail()
```

```python
print(data.shape)
print(data.dtypes)
```

```python
data = data.apply(pd.to_numeric)
data.dtypes
```

```python
data.describe()
```

```python
data.hist(figsize = (15, 15))
plt.show()
```

```python
# create X and Y datasets for training
from sklearn.model_selection import train_test_split

X = np.array(data.drop(['class'], axis=1))
y = np.array(data['class'])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=42)
```

```python
# convert the data to categorical labels
from keras.utils.np_utils import to_categorical

y_train_cat = to_categorical(y_train, num_classes=None)
y_test_cat = to_categorical(y_test, num_classes=None)
print(y_train_cat.shape)
print(y_train_cat[:10])
```

### Building and Training the Neural Network

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam

def create_model():
    # model
    model = Sequential([
    Dense(8, input_dim=13, kernel_initializer='normal', activation='relu'),
    Dense(4, kernel_initializer='normal', activation='relu'),
    Dense(5, activation='softmax'),
    ])
    # compiling model
    adam = Adam(lr=0.001)
    model.compile(loss='categorical_crossentropy', optimizer=adam, metrics=['accuracy'])
    return model

# model summary
model = create_model()
print(model.summary())
```

```python
model.fit(X_train,y_train_cat,epochs=100,batch_size=10,verbose=1)
```

```python
from sklearn.metrics import classification_report, accuracy_score

categorical_pred = model.predict_classes(X_test)
print(categorical_pred)
print('Results for Categorical Model')
print(accuracy_score(y_test, categorical_pred))
print(classification_report(y_test, categorical_pred))
```

```python
# convert into binary classification problem - heart disease or no heart disease
data["class"] = data["class"].apply(lambda x: 1 if x >= 1 else 0)
X_bin=data.drop("class",axis=1)
y_bin=data["class"]
```

```python
X_train_bin, X_test_bin, y_train_bin, y_test_bin = train_test_split(X_bin, y_bin, test_size = 0.3, random_state=42)
```

```python
def create_binary_model():
    # model
    model = Sequential([
    Dense(8, input_dim=13, kernel_initializer='normal', activation='relu'),
    Dense(4, kernel_initializer='normal', activation='relu'),
    Dense(1, activation='sigmoid'),
    ])
    # compiling model
    adam = Adam(lr=0.001)
    model.compile(loss='binary_crossentropy', optimizer=adam, metrics=['accuracy'])
    return model

# model summary
binary_model = create_binary_model()
print(binary_model.summary())
```

```python
binary_model.fit(X_train_bin, y_train_bin, epochs=150, batch_size=10, verbose = 1)
```

```python
from sklearn.metrics import classification_report, accuracy_score
binary_pred = np.round(binary_model.predict(X_test)).astype(int)

print('Results for Binary Model')
print(accuracy_score(y_test_bin, binary_pred))
print(classification_report(y_test_bin, binary_pred))
```
